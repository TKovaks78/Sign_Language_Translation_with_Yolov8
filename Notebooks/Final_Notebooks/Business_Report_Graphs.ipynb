# 1. Setup

```python
# Import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the training results dataframes

# Results for baseline with freezing
df_freezing = pd.read_csv("/content/With_freezing_results.csv")

# Results for baseline without freezing
df_no_freezing = pd.read_csv("/content/No_freezing_results.csv")

# Results for baseline with Adam Optimizer
df_Adam = pd.read_csv("/content/Adam_results.csv")

# Results for baseline with AdamW Optimizer
df_AdamW = pd.read_csv("/content/AdamW_results.csv")

# Results for baseline with NAG Optimizer
df_NAG = pd.read_csv("/content/NAG_results.csv")

# Results for data augmentation
df_aug = pd.read_csv("/content/data_aug_results.csv")

# Results for hyperparameters 1
df_hp1 = pd.read_csv("/content/hp1_results.csv")

# Results for hyperparameters 2
df_hp2 = pd.read_csv("/content/hp2_results.csv")
```

```python
# Display one results dataframe
df_Adam.head(5)
```

```python
print(df_Adam.columns)
```

```python
# Create list with the columns we want to plot
col_list = df_Adam.columns[4:11].tolist()
col_list
```

# 2. Transfer Learning

```python
# Define a function to plot the graphs

def plot_freeze(col_name):
  # Set the x-axis data
  x = df_freezing['                  epoch']

  # Set the y-axis data for each dataframe
  y_freezing = df_freezing[col_name]
  y_no_freezing = df_no_freezing[col_name]

  # Create the graph
  plt.plot(x, y_no_freezing, color='red', label='no freezing')
  plt.plot(x, y_freezing, color='blue', label='with freezing')

  # Add labels and legend
  plt.xlabel('Epoch')
  plt.ylabel(col_name)
  plt.legend()

  # Show the graph
  plt.show()
```

```python
# Plot the results
for col in col_list:
  plot_freeze(col)
```

# 3. Optimizers

```python
# Define a function to plot the graphs

def plot_opt(col_name):
  # Set the x-axis data
  x = df_Adam['                  epoch']

  # Set the y-axis data for each dataframe
  y_Adam = df_Adam[col_name]
  y_AdamW = df_AdamW[col_name]
  y_NAG = df_NAG[col_name]

  # Create the graph
  plt.plot(x, y_Adam, color='red', label='Adam')
  plt.plot(x, y_AdamW, color='green', label='AdamW')
  plt.plot(x, y_NAG, color='blue', label='NAG')

  # Add labels and legend
  plt.xlabel('Epoch')
  plt.ylabel(col_name)
  plt.legend()

  # Show the graph
  plt.show()
```

```python
# Plot the results
for col in col_list:
  plot_opt(col)
```

# 4. Data Augmentation

```python
# Define a function to plot the graphs

def plot_aug(col_name):
  # Set the x-axis data
  x = df_freezing['                  epoch']

  # Set the y-axis data for each dataframe
  y_freezing = df_freezing[col_name]
  y_aug = df_aug[col_name]

  # Create the graph
  plt.plot(x, y_freezing, color='red', label='baseline')
  plt.plot(x, y_aug, color='blue', label='augmented data')

  # Add labels and legend
  plt.xlabel('Epoch')
  plt.ylabel(col_name)
  plt.legend()

  # Show the graph
  plt.show()
```

```python
# Plot the results
for col in col_list:
  plot_aug(col)
```

# 5. Hyperparameters

```python
# Results for data augmentation
df_aug = pd.read_csv("/content/data_aug_results.csv")

# Results for hyperparameters 1
df_hp1 = pd.read_csv("/content/hp1_results.csv")

# Results for hyperparameters 2
df_hp2 = pd.read_csv("/content/hp2_results.csv")
```

```python
# Create list with the columns we want to plot
col_list = df_aug.columns[4:11].tolist()
col_list
```

```python
# Define a function to plot the graphs

def plot_hp(col_name):
  # Set the x-axis data
  x = df_aug['                  epoch'].iloc[:93]

  # Set the y-axis data for each dataframe
  y_aug = df_aug[col_name].iloc[:93]
  y_hp1 = df_hp1[col_name].iloc[:93]
  y_hp2 = df_hp2[col_name].iloc[:93]

  # Create the graph
  plt.plot(x, y_aug, color='red', label='post-data augmentation')
  plt.plot(x, y_hp1, color='green', label='1. Lr = 0.0001, dropout = 0.01')
  plt.plot(x, y_hp2, color='blue', label='2. beta = 0.9')

  # Add labels and legend
  plt.xlabel('Epoch')
  plt.ylabel(col_name)
  plt.legend()

  # Show the graph
  plt.show()
```

```python
# Plot the results
for col in col_list:
  plot_hp(col)
```
